```
import numpy as np
import matplotlib.pyplot as plt
```

```
x = np.arange(0, 4 * np.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label='sin')
plt.plot(x, y2, label='cos')
plt.legend()
plt.show()
```

```
x = np.arange(0, 4 * np.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(2, 1, 1)
plt.plot(x, y1, label='sin')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(x, y2, label='cos')
plt.legend()

plt.show()
```
